```python
arr=dict()
for i in range(23):
    arr[target1.axes[0][i]]=target1['target'][i]
grouped_target=pd.DataFrame(list(arr.items()),columns=['target','count'])
grouped_target
```

Adding another column in the grouped target so that we can categrazied these attack types to five major classes:
1. NORMAL
2. DOS
3. U2R
4. R2L
5. PROBE

adding group accordings:
NORMAL= ['normal'] <br/>
DOS=['neptune','back','teardrop','smurf','pod','land'] <br/>
R2L=['warezclient','multihop','ftp_write','imap','guess_passwd','warezmaster','spy','phf']<br/>
U2R=['buffer_overflow','loadmodule','rootkit','perl']<br/>
PROBE=['portsweep','satan','ipsweep','nmap'] <br/>


```python
NORMAL= ['normal'] 
DOS=['neptune','back','teardrop','smurf','pod','land'] 
R2L=['warezclient','multihop','ftp_write','imap','guess_passwd','warezmaster','spy','phf']
U2R=['buffer_overflow','loadmodule','rootkit','perl']
PROBE=['portsweep','satan','ipsweep','nmap'] 
ans=[]
for i in grouped_target['target']:
    if(i in DOS):
        ans.append('DOS')
    elif(i in R2L):
        ans.append('R2L')
    elif(i in U2R):
        ans.append('U2R')
    elif(i in PROBE):
        ans.append('PROBE')
    elif(i in NORMAL):
        ans.append('NORMAL')
print(ans)
grouped_target['category']=ans
grouped_target
```

```python

grouped_target.groupby(['category']).sum()
```

```python
groupes=grouped_target.groupby(['category']).sum()
```

```python
x_axis=np.array(groupes.axes[0])
y_axis=np.array(groupes['count'])
print(x_axis)
print(y_axis)
##This show how to convert group by to normal numpy array
plt.figure(figsize=(10,7))
g=sns.barplot(x=x_axis,y=y_axis)
plt.ylim([0,450000])
plt.xlabel('Types of Attack')
plt.ylabel('Count')
for i in range(5):
    g.text(i,y_axis[i]+10000,y_axis[i])
    g.text(i,y_axis[i]+25000,str(round((y_axis[i]/df.shape[0])*100,4))+'%')
plt.show()
```

```python
df['protocol_type']=df['protocol_type'].apply(lambda x: x[2:-1])
df.groupby(['protocol_type','target'])['target'].count()
```

```python
plt.figure(figsize=(10,7))
g=sns.countplot(x='protocol_type',data=df)
plt.ylim([0,350000])
for p in g.patches:
        g.annotate(p.get_height(), (p.get_x()+0.3, p.get_height()+5000))
for p in g.patches:
        g.annotate('{:.3f}%'.format((p.get_height()/df.shape[0])*100), (p.get_x()+0.3, p.get_height()+15000))
plt.show()
```

```python
NORMAL= ['normal'] 
DOS=['neptune','back','teardrop','smurf','pod','land'] 
R2L=['warezclient','multihop','ftp_write','imap','guess_passwd','warezmaster','spy','phf']
U2R=['buffer_overflow','loadmodule','rootkit','perl']
PROBE=['portsweep','satan','ipsweep','nmap'] 
ans=[]
for i in df['target']:
    if(i in DOS):
        ans.append('DOS')
    elif(i in R2L):
        ans.append('R2L')
    elif(i in U2R):
        ans.append('U2R')
    elif(i in PROBE):
        ans.append('PROBE')
    elif(i in NORMAL):
        ans.append('NORMAL')
print(ans)
df['category']=ans
df.head()
```

## Attack types vs Protocols used


```python
cont=0
plt.figure(figsize=(10,7))
g=sns.countplot(x='category',hue='protocol_type',data=df, order=df['category'].value_counts().index)
plt.ylim([0,300000])
for p in g.patches:
    y=p.get_height()
    if np.isnan(y):
        y=0
    g.annotate(str(np.int(y))+(' ({:.3f}%)'.format((y/y_axis[cont%5])*100)), (p.get_x()+0.1, 100000),rotation=90)
    cont=cont+1
plt.legend(loc='upper right')
plt.xlabel('Types of attack')
plt.show()
```
