```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data=np.array([['Naive Bayes','label','train',98.318],
               ['Naive Bayes','label','test',98.418],
               ['Naive Bayes','category','train',87.921],
               ['Naive Bayes','category','test',88.058],
               ['Naive Bayes','target','train',89.686],
               ['Naive Bayes','target','test',89.658],
               ['Logistic','label','train',99.344],
               ['Logistic','label','test',99.368],
               ['Logistic','category','train',99.430],
               ['Logistic','category','test',99.454],
               ['Logistic','target','train',99.462],
               ['Logistic','target','test',99.491],
               ['SVM','label','train',99.2939],
               ['SVM','label','test',99.3361],
               ['SVM','category','train',99.3378],
               ['SVM','category','test',99.3624],
               ['SVM','target','train',99.3462],
               ['SVM','target','test',99.3765],
               ['Random Forest','label','train',99.99682],
               ['Random Forest','label','test',99.97234],
               ['Random Forest','category','train',99.99653],
               ['Random Forest','category','test',99.96694],
               ['Random Forest','target','train',99.99682],
               ['Random Forest','target','test',99.96491]
              ])
```

```python
data
```

```python
df=pd.DataFrame(data,columns=['algorithms','type','split','score'])
df['score']=df['score'].astype('float')
```

```python
df.head()
```

```python
df[df['type']=='label']
```

```python
plt.figure(figsize=(10,7))
g=sns.barplot(x='algorithms',y='score',hue='split',data=df[df['type']=='label'])
for p in g.patches:
        g.annotate('{:.4f}'.format(p.get_height()), (p.get_x()+0.1, p.get_height()+0.0001))
plt.ylim([96,101])
plt.show()
```

```python
plt.figure(figsize=(10,7))
g=sns.barplot(x='algorithms',y='score',hue='split',data=df[df['type']=='category'])
for p in g.patches:
        g.annotate('{:.4f}'.format(p.get_height()), (p.get_x()+0.1, p.get_height()+0.0001))
plt.ylim([85,101])
plt.title('Accuray of different algorithms on categorised attack ')
plt.show()
```

```python
plt.figure(figsize=(10,7))
g=sns.barplot(x='algorithms',y='score',hue='type',data=df[df['split']=='test'])
for p in g.patches:
        g.annotate('{:.4f}'.format(p.get_height()), (p.get_x()+0.1, p.get_height()+2.5),rotation=90)
#g.annotate(rotation=90)
plt.ylim([85,110])
plt.title('Accuray of different algorithms on test data')
plt.show()
```
