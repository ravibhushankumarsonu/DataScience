```python
plt.figure(figsize=(15,15))
plt.plot(x,Data.loc[(Data['CD Block_Code']==400)&(Data['Level']=='VILLAGE')]['Total Population Person'],'-b')
plt.plot(x,Data.loc[(Data['CD Block_Code']==400)&(Data['Level']=='VILLAGE')]['Total Population Male'],'-g')
plt.plot(x,Data.loc[(Data['CD Block_Code']==400)&(Data['Level']=='VILLAGE')]['Total Population Female'],'-r')
plt.xticks(x,Data.loc[(Data['CD Block_Code']==400)&(Data['Level']=='VILLAGE')]['Name'],rotation=90)
plt.grid()
plt.show()
```

```python
plt.figure(figsize=(15,15))
sns.barplot(x=Data.loc[(Data['Level']=='CD BLOCK')&(Data['Total/Rural/Urban']=='Total')]['Name'],
            y=Data.loc[(Data['Level']=='CD BLOCK')]['Total Population Person'],
            data=Data)
plt.xticks(rotation=90)
plt.show()
```

```python
blocks=np.linspace(0,len(Data.loc[(Data['Level']=='CD BLOCK')&(Data['Total/Rural/Urban']=='Total')]['Name'])-1,
                   len(Data.loc[(Data['Level']=='CD BLOCK')&(Data['Total/Rural/Urban']=='Total')]['Name']))
print(blocks)
```

```python
plt.figure(figsize=(15,15))
plt.plot(blocks,Data.loc[(Data['Level']=='CD BLOCK')&(Data['Total/Rural/Urban']=='Total')]['Total Population Person'],'-b')
plt.plot(blocks,Data.loc[(Data['Level']=='CD BLOCK')&(Data['Total/Rural/Urban']=='Total')]['Total Population Male'],'-g')
plt.plot(blocks,Data.loc[(Data['Level']=='CD BLOCK')&(Data['Total/Rural/Urban']=='Total')]['Total Population Female'],'-r')
plt.xticks(blocks,Data.loc[(Data['Level']=='CD BLOCK')&(Data['Total/Rural/Urban']=='Total')]['Name'],rotation=90)
plt.grid()
plt.show()
```

```python
plt.figure(figsize=(15,15))
plt.plot(blocks,Data.loc[(Data['Level']=='CD BLOCK')&(Data['Total/Rural/Urban']=='Total')]['Total Population Person'],'-b')
plt.plot(blocks,Data.loc[(Data['Level']=='CD BLOCK')&(Data['Total/Rural/Urban']=='Total')]['Literates Population Person'],'-g')
#plt.plot(blocks,Data.loc[(Data['Level']=='CD BLOCK')&(Data['Total/Rural/Urban']=='Total')]['Total Population Female'],'-r')
plt.xticks(blocks,Data.loc[(Data['Level']=='CD BLOCK')&(Data['Total/Rural/Urban']=='Total')]['Name'],rotation=90)
plt.grid()
plt.show()
```

```python
plt.figure(figsize=(15,15))
plt.plot(blocks,Data.loc[(Data['Level']=='CD BLOCK')&(Data['Total/Rural/Urban']=='Total')]['Total Population Male'],'-b')
plt.plot(blocks,Data.loc[(Data['Level']=='CD BLOCK')&(Data['Total/Rural/Urban']=='Total')]['Literates Population Male'],'-g')
#plt.plot(blocks,Data.loc[(Data['Level']=='CD BLOCK')&(Data['Total/Rural/Urban']=='Total')]['Total Population Female'],'-r')
plt.xticks(blocks,Data.loc[(Data['Level']=='CD BLOCK')&(Data['Total/Rural/Urban']=='Total')]['Name'],rotation=90)
plt.grid()
plt.show()
```

```python
plt.figure(figsize=(15,15))
plt.plot(blocks,Data.loc[(Data['Level']=='CD BLOCK')&(Data['Total/Rural/Urban']=='Total')]['Total Population Female'],'-b')
plt.plot(blocks,Data.loc[(Data['Level']=='CD BLOCK')&(Data['Total/Rural/Urban']=='Total')]['Literates Population Female'],'-g')
#plt.plot(blocks,Data.loc[(Data['Level']=='CD BLOCK')&(Data['Total/Rural/Urban']=='Total')]['Total Population Female'],'-r')
plt.xticks(blocks,Data.loc[(Data['Level']=='CD BLOCK')&(Data['Total/Rural/Urban']=='Total')]['Name'],rotation=90)
plt.grid()
plt.show()
```

```python
plt.figure(figsize=(15,15))
val=np.arange(len(Data.loc[(Data['Level']=='CD BLOCK')&(Data['Total/Rural/Urban']=='Total')]['Name']))
#print(ax)
bar_width=0.35
opacity=0.8
plt.bar(val,Data.loc[(Data['Level']=='CD BLOCK')&(Data['Total/Rural/Urban']=='Total')]['Total Population Female'],
        bar_width,color='b', alpha=opacity,label='Total Population Female')
plt.bar(val+bar_width,Data.loc[(Data['Level']=='CD BLOCK')&(Data['Total/Rural/Urban']=='Total')]['Literates Population Female'],
         bar_width,color='r',alpha=opacity,label='Literates Population Female')
#plt.plot(blocks,Data.loc[(Data['Level']=='CD BLOCK')&(Data['Total/Rural/Urban']=='Total')]['Total Population Female'],'-r')
plt.xticks(val+bar_width,Data.loc[(Data['Level']=='CD BLOCK')&(Data['Total/Rural/Urban']=='Total')]['Name'],rotation=90)
#plt.grid()
plt.legend()
plt.show()
```
