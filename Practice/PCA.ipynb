Comparison of LDA and PCA 2D projection of Iris dataset

```python
print(__doc__)
import matplotlib.pyplot as plt
from sklearn import datasets
from sklearn.decomposition import PCA
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
```

```python
Data=datasets.load_iris()
```

```python
X=Data.data
Y=Data.target
```

```python
X.shape
```

```python
target_names=Data.target_names
print(target_names)
```

```python
#Y
```

```python
pca=PCA(n_components=2)
x_r=pca.fit(X).transform(X)
```

```python
#x_r
```

```python
lda=LinearDiscriminantAnalysis(n_components=2)
x_r1=lda.fit(X,Y).transform(X)
```

```python
#x_r1
```

```python
print('Explained variance of PCA %s'% str(pca.explained_variance_ratio_))
```

```python
plt.figure()
colors=['navy', 'turquoise', 'darkorange']
lw=2
```

```python
for color, i, target_name in zip(colors, [0,1,2],target_names):
    plt.scatter(x_r[Y==i,0],x_r[Y==i,1],color=colors,alpha=0.8,lw=lw,label=target_name)
plt.legend(loc='best', shadow=False, scatterpoints=1)
plt.title('PCA of Iris dataset')
plt.show()
```

```python
plt.figure()
for color, i, target_name in zip(colors, [0,1,2],target_names):
    plt.scatter(x_r1[Y==i,0],x_r1[Y==i,1],color=colors,alpha=0.8,lw=lw,label=target_name)
plt.legend(loc='best', shadow=False, scatterpoints=1)
plt.title('LDA of Iris dataset')
plt.show()
```

```python
'''
print(__doc__)

import matplotlib.pyplot as plt

from sklearn import datasets
from sklearn.decomposition import PCA
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
'''
iris = datasets.load_iris()

X = iris.data
y = iris.target
target_names = iris.target_names

pca = PCA(n_components=2)
X_r = pca.fit(X).transform(X)

lda = LinearDiscriminantAnalysis(n_components=2)
X_r2 = lda.fit(X, y).transform(X)

# Percentage of variance explained for each components
print('explained variance ratio (first two components): %s'
      % str(pca.explained_variance_ratio_))

plt.figure()
colors = ['navy', 'turquoise', 'darkorange']
lw = 2

for color, i, target_name in zip(colors, [0, 1, 2], target_names):
    plt.scatter(X_r[y == i, 0], X_r[y == i, 1], color=color, alpha=.8, lw=lw,
                label=target_name)
plt.legend(loc='best', shadow=False, scatterpoints=1)
plt.title('PCA of IRIS dataset')

plt.figure()
for color, i, target_name in zip(colors, [0, 1, 2], target_names):
    plt.scatter(X_r2[y == i, 0], X_r2[y == i, 1], alpha=.8, color=color,
                label=target_name)
plt.legend(loc='best', shadow=False, scatterpoints=1)
plt.title('LDA of IRIS dataset')

plt.show()
```
