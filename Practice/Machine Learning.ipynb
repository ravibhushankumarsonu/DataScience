```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
z=np.linspace(-5,5,100)
```

```python
x=1/(1+np.e**(-z))
plt.plot(z,x)
plt.xlabel('z')
plt.ylabel('value')
plt.title('Sigmoid Function')
plt.show()
```
