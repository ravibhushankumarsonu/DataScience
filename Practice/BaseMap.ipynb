>Basemap is the matplotlib utility library for the ploting on the map. It's help to plot projection on the Geographic Image of World.

```python
from mpl_toolkits.basemap import Basemap
import numpy as np
import matplotlib.pyplot as plt

```

***BaseMap*** provides different types of projection

```python
plt.figure(figsize=(20,10))
map=Basemap()
map.drawcoastlines()
plt.show()
```

```python
plt.figure(figsize=(20,10))
map = Basemap(projection='ortho', 
              lat_0=21, lon_0=78)
#Fill the globe with a blue color 
map.drawmapboundary(fill_color='aqua')
#Fill the continents with the land color
map.fillcontinents(color='coral',lake_color='aqua')

map.drawcoastlines()
plt.show()
```

```python
plt.figure(figsize=(20,10))
map = Basemap(projection='merc', llcrnrlat=5.8, urcrnrlat=38,
                llcrnrlon=63, urcrnrlon=93, resolution='h')
#Fill the globe with a blue color 
map.drawmapboundary()
#Fill the continents with the land color
#map.fillcontinents(color='coral',lake_color='aqua')
map.drawcountries()
#map.drawstates()
map.drawcoastlines()
map.drawrivers()
plt.show()
```
