Make the predication model


```python
plt.figure(figsize=(15,15))
plt.scatter(Data['price'],Data['bathrooms'])
plt.grid()
plt.xticks(rotation=90)
plt.show()
```

```python
from sklearn import linear_model
#regr=linear_model.LinearRegression().fit(train['sqft_living'],train['price'])
x=train['sqft_living'].values
y=train['price'].values
x=x.reshape(len(x),1)
y=y.reshape(len(y),1)
```

```python
regr=linear_model.LinearRegression().fit(x,y)
```

```python
print(regr.coef_)
```

```python
print(x)
```

```python
#plt.figure(figsize=(15,15))
plt.scatter(x,y,color='black')
plt.plot(x,regr.predict(x),color='green')
plt.grid()
plt.xticks(rotation=90)
plt.show()
```
