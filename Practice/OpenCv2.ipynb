```python
url='http://s0.geograph.org.uk/photos/40/57/405725_b17937da.jpg'
url_response=requests.get(url)
image_arr=np.array(bytearray(url_response.content),dtype=np.uint8)
img=cv2.imdecode(image_arr,-1)
plt.figure(figsize=(10,10))
cv2.rectangle(img,(15,25),(200,150),(34,34,25),20)
plt.imshow(img)
plt.show()
```

Loading the video source from the opencv press q to quite the open windows

```python
cap = cv2.VideoCapture(0)
while(True):
    ret, frame = cap.read()
    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
 
    cv2.imshow('frame',gray)
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()
```
