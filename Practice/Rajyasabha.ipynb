```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import datetime as dt
```

```python
df=pd.read_csv('Rajyasabha_QnA.csv')
```

```python
df.info()
```

```python
df.head()
```

```python
print(len(df))
```

```python
plt.figure(figsize=(20,10))
sns.countplot(x='question_by',data=df[df.answer_date=='11.08.2017'])
plt.xticks(rotation=90)
plt.show()
```

```python
df['answer_date']=pd.to_datetime(df['answer_date'])
```

```python
print(df.answer_date.dtypes)
```

```python
months=['Jan','Feb','Mar','April','May','June','July','Aug','Sept','Oct','Nov','Dec']
plt.figure(figsize=(10,10))
ax=sns.countplot(x=df['answer_date'].dt.month,hue=df['answer_date'].dt.year,data=df)
plt.xticks(rotation=90)
ax.set(xticklabels=months)
for p in ax.patches:
    ax.annotate('{:.0f}'.format(p.get_height()),(p.get_x()+0.1,p.get_y()+p.get_height()+130),rotation=90)
plt.show()
```

```python
plt.figure(figsize=(20,20))
ax=sns.countplot(x=df['ministry'],data=df)
plt.xticks(rotation=90)
#ax.set(xticklabels=months)
for p in ax.patches:
    ax.annotate('{:.0f}'.format(p.get_height()),(p.get_x()+0.1,p.get_y()+p.get_height()+30),rotation=90)
plt.show()
```

```python
plt.figure(figsize=(20,20))
ax=sns.countplot(x=df.loc[df['answer_date'].dt.year==2017]['ministry'],data=df)
#ax=sns.countplot(x=df.loc[df['answer_date'].dt.year==2017]['ministry'],data=df)
plt.xticks(rotation=90)
#ax.set(xticklabels=months)
for p in ax.patches:
    ax.annotate('{:.0f}'.format(p.get_height()),(p.get_x()+0.1,p.get_y()+p.get_height()+10),rotation=90)
plt.show()
```
